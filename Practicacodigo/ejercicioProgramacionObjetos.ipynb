```python
# Clase llamada estudiante
class Estudiante:
    def __init__(self,nombre,edad,curso,nota):
        self.nombre = nombre
        self.edad = edad
        self.curso = curso
        self.nota = nota

    def mostrar_datos(self):
        print("Mi nombre es", self.nombre)
        print("Mi edad es", self.edad,"años")
        print("Estoy en el curso", self.curso)
        print("Mi nota es", self.nota)

    def aprobo(self):
        if self.nota >= 3:
            print("Aprobo 👍") 
        else:
            print("reprobo ❌")   


mateo = Estudiante("Mateo", 14, "decimo B", 4)  
mateo.mostrar_datos()
mateo.aprobo()





        
```
