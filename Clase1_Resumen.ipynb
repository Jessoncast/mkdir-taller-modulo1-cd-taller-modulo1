## 1. IA en el mundo real: un caso concreto

En clase vimos qué es la IA y entrenamos un modelo de visión. Acá vamos a ver otro dominio completamente distinto: **el diagnóstico médico**.

### ¿Cómo diagnostica un médico vs. cómo diagnostica un modelo de IA?

**El médico (razonamiento humano):**
Mira una radiografía. Basándose en años de experiencia, reconoce patrones que aprendió viendo miles de casos. Dice: *"Esta sombra en el lóbulo derecho, combinada con estos síntomas, me indica neumonía."* Puede explicar su razonamiento.

**El modelo de IA (razonamiento estadístico):**
Fue entrenado con 100.000 radiografías etiquetadas por médicos. Aprendió qué combinaciones de píxeles se asocian con cada diagnóstico. Dice: *"91.3% de probabilidad de neumonía."* No puede explicar por qué — solo sabe que ese patrón de píxeles apareció junto a ese diagnóstico miles de veces.

📐Esto se llama el problema de la **interpretabilidad** o *explainability* en IA. Un modelo puede ser muy preciso pero completamente opaco — no sabés por qué tomó esa decisión. En medicina, eso es un problema ético y legal grave. En publicidad, menos.

💬El médico puede señalar la radiografía y decirte *"mirá acá"*. El modelo te da un número pero no puede señalar nada. Ambos aciertan — pero uno puede rendir cuentas y el otro no. Ese es uno de los debates más importantes de la IA hoy.

### El modelo más simple posible: detección de spam

Antes de entrenar modelos complejos, mirá cómo funciona la lógica básica de clasificación.
Este código **no usa IA** — usa reglas escritas a mano. Pero la *estructura del problema* es idéntica:

````python
# Clasificador de spam sin IA — solo lógica
# Esto es lo que hacíamos ANTES del Machine Learning

def es_spam(email):
    """
    Analiza un email y decide si es spam.
    Devuelve True si es spam, False si es legítimo.
    """
    email_minuscula = email.lower()
    
    palabras_sospechosas = [
        "ganaste", "gratis", "urgente", "clic aquí",
        "oferta exclusiva", "millones", "premio"
    ]
    
    for palabra in palabras_sospechosas:
        if palabra in email_minuscula:
            return True, f"Encontré la palabra sospechosa: '{palabra}'"
    
    return False, "Sin palabras sospechosas"


# Emails de prueba
emails = [
    "Hola, ¿podemos reagendar la reunión del jueves?",
    "¡GANASTE un iPhone! Hacé clic aquí para reclamar tu PREMIO GRATIS",
    "Adjunto el informe que me pediste. Saludos.",
    "URGENTE: Tu cuenta será suspendida. Entrá ya."
]

print("Analizando emails...\n")
for email in emails:
    resultado, razon = es_spam(email)
    etiqueta = "🚨 SPAM" if resultado else "✅ LEGÍTIMO"
    print(f"{etiqueta}")
    print(f"  Email:  {email[:55]}..." if len(email) > 55 else f"  Email:  {email}")
    print(f"  Razón:  {razon}\n")
````

### ¿Por qué este enfoque tiene un límite?

Ejecutaste el código. Funciona. ¿Entonces cuál es el problema?

El problema es el **mantenimiento**. Un spammer solo necesita cambiar "ganaste" por "eres el elegido" y el filtro no lo detecta. Entonces el ingeniero agrega esa palabra. El spammer cambia de nuevo. Es una carrera infinita.

**Con Machine Learning**, el modelo aprende de miles de emails nuevos continuamente. No necesita que un humano actualice las reglas — las infiere solo de los datos. **Ese** es el salto que hace el ML.

## 3. Data Science: los datos mienten si no los entendés

En clase vimos que sin buenos datos no hay buena IA. Acá vamos a ver algo más sutil: **datos correctos que llevan a conclusiones incorrectas**.

### La Paradoja de Simpson

Un hospital analiza dos tratamientos para una enfermedad:

| | Tratamiento A | Tratamiento B |
|---|---|---|
| Pacientes leves | 81 de 87 curados (93%) | 234 de 270 curados (87%) |
| Pacientes graves | 192 de 263 curados (73%) | 55 de 80 curados (69%) |
| **Total** | **273 de 350 curados (78%)** | **289 de 350 curados (83%)** |

**El tratamiento A es mejor en pacientes leves. Es mejor en pacientes graves. Pero en total, parece que B es mejor.**

Esto es real. Ocurre porque B recibió más pacientes leves (más fáciles de curar). Un modelo de IA entrenado con el total sin entender esto recomendaría B — y estaría equivocado para cualquier paciente individual.

📐La Paradoja de Simpson es un fenómeno estadístico donde una tendencia que aparece en grupos separados desaparece o se invierte cuando los grupos se combinan. Sucede por variables de confusión no controladas.

💬Amigable: Los números no mienten, pero el contexto sí puede faltar. Antes de darle datos a un modelo, un Data Scientist tiene que entender qué hay detrás de esos números — quién los generó, bajo qué condiciones, qué está faltando. Por eso el análisis exploratorio (EDA) existe antes del modelado.

### — Variables y tipos de datos

📐Una variable es un nombre que apunta a una zona de memoria donde se almacena un valor. Python infiere el tipo automáticamente — no necesitás declararlo.

💬Una variable es una caja con una etiqueta. La etiqueta es el nombre, la caja guarda el valor. Podés cambiar lo que hay adentro en cualquier momento.


````python
# Los 5 tipos de datos que más vas a usar

nombre = "Ana"           # str  — texto
edad = 25                # int  — número entero
altura = 1.75            # float — número decimal
es_estudiante = True     # bool — verdadero o falso
notas = [8, 9, 7, 10]   # list — colección ordenada

print(f"Nombre: {nombre} → tipo: {type(nombre).__name__}")
print(f"Edad: {edad} → tipo: {type(edad).__name__}")
print(f"Altura: {altura} → tipo: {type(altura).__name__}")
print(f"Estudiante: {es_estudiante} → tipo: {type(es_estudiante).__name__}")
print(f"Notas: {notas} → tipo: {type(notas).__name__}")
````

### Strings: el tipo de dato del lenguaje natural

En Data Science y NLP (Procesamiento de Lenguaje Natural) vas a trabajar muchísimo con texto. Conocer los métodos básicos de strings es fundamental.


````python
frase = "Inteligencia Artificial y Ciencia de Datos"

print(f"Original:      {frase}")
print(f"Mayúsculas:    {frase.upper()}")
print(f"Minúsculas:    {frase.lower()}")
print(f"Longitud:      {len(frase)} caracteres")
print(f"Reemplazar:    {frase.replace('Artificial', 'General')}")
print(f"¿Empieza con 'IA'?: {frase.startswith('Inteligencia')}")

# Acceso por índice (los índices empiezan en 0)
print(f"\nPrimer carácter:  {frase[0]}")
print(f"Últimos 5 chars:  {frase[-5:]}")
print(f"Primeras 12:      {frase[:12]}")

# f-strings: la forma moderna de formatear texto
nombre = "Martín"
modulo = 1
print(f"\nHola {nombre}, estás en el Módulo {modulo} del máster.")
````

### Listas: colecciones ordenadas

📐Una lista es una secuencia mutable y ordenada de elementos. Puede contener tipos heterogéneos. Acceso por índice en O(1).

💬Una lista es como una fila de cajitas numeradas. Podés poner cualquier cosa en cada cajita, en cualquier orden, y cambiarlas cuando quieras. En Data Science, los datasets son esencialmente listas de listas.


````python
# Listas en acción
tecnologias_ia = ["Python", "TensorFlow", "PyTorch", "scikit-learn", "Pandas"]

print(f"Lista completa: {tecnologias_ia}")
print(f"Primera: {tecnologias_ia[0]}")
print(f"Última:  {tecnologias_ia[-1]}")
print(f"Cantidad: {len(tecnologias_ia)} tecnologías")

# Modificar
tecnologias_ia.append("Keras")         # agregar al final
tecnologias_ia.insert(1, "NumPy")      # insertar en posición
tecnologias_ia.remove("TensorFlow")    # eliminar por valor

print(f"\nLista actualizada: {tecnologias_ia}")

# Slicing — extraer porciones
print(f"Primeras 3: {tecnologias_ia[:3]}")
print(f"Del índice 2 al 4: {tecnologias_ia[2:4]}")

# Verificar existencia
print(f"\n¿Está 'Python'?: {'Python' in tecnologias_ia}")
````

### Diccionarios: datos con estructura

📐Un diccionario es una tabla hash que almacena pares clave-valor. Acceso por clave en O(1). Las claves deben ser inmutables y únicas.

💬Un diccionario es como una ficha técnica. En vez de acceder por número (como en la lista), accedés por nombre. En Data Science, cada fila de un dataset se puede pensar como un diccionario: columna → valor.


````python
# Un alumno del máster representado como diccionario
alumno = {
    "nombre": "Laura",
    "edad": 28,
    "ciudad": "Buenos Aires",
    "experiencia_python": False,
    "modulos_completados": []
}

# Acceso
print(f"Nombre: {alumno['nombre']}")
print(f"Ciudad: {alumno['ciudad']}")

# .get() — acceso seguro (no lanza error si la clave no existe)
print(f"Profesión: {alumno.get('profesion', 'No especificada')}")

# Modificar y agregar
alumno['experiencia_python'] = True
alumno['modulos_completados'].append('Módulo 1')
alumno['linkedin'] = 'linkedin.com/in/laura'

print(f"\nAlumno actualizado:")
for clave, valor in alumno.items():
    print(f"  {clave}: {valor}")
````

### Un sistema de recomendación básico (todo junto)

Este ejemplo usa todo lo que aprendiste — listas, diccionarios, strings — para construir algo que se parece a lo que hace Spotify o Netflix en su versión más simple posible.


````python
# Sistema de recomendación simple
# (Versión real usa ML — esto es la lógica fundamental)

catalogo = {
    'accion': ['Matrix', 'John Wick', 'Mad Max: Fury Road'],
    'comedia': ['Superbad', 'The Hangover', 'Game Night'],
    'drama':   ['The Shawshank Redemption', 'Forrest Gump', 'Parasite'],
    'ciencia ficcion': ['Interstellar', 'Blade Runner 2049', 'Ex Machina']
}

def recomendar(genero_favorito):
    """
    Recomienda una película según el género.
    .lower() normaliza el input — igual funciona 'ACCION' que 'accion'.
    """
    genero = genero_favorito.lower()
    
    if genero in catalogo:
        pelicula = catalogo[genero][0]
        return f"🎬 Recomendación para '{genero}': {pelicula}"
    else:
        disponibles = ', '.join(catalogo.keys())
        return f"Género no encontrado. Géneros disponibles: {disponibles}"

# Prueba
print(recomendar('Accion'))
print(recomendar('drama'))
print(recomendar('ciencia ficcion'))
print(recomendar('terror'))  # género que no existe
````
